```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv("forbes_billionaires000.csv")
```

```python
# Set visualization styles
sns.set_theme(style="whitegrid")
plt.rcParams.update({'axes.titlesize': 14, 'axes.labelsize': 12})
```

```python
### 1. lklr gr milyaderlr (Top 20) ###
plt.figure(figsize=(12, 6))
top_countries = df['Country'].value_counts().head(20)
sns.barplot(y=top_countries.index, x=top_countries.values, palette='viridis')
plt.xlabel("Milyaderlrin Say")
plt.ylabel("lk")
plt.title("Milyader sayna gr top 20 lk")
plt.show()
```

```python
### 2. Milyarderlrin Ya Paylanmas ###
plt.figure(figsize=(12, 6))
sns.histplot(df["Age"].dropna(), bins=20, kde=True, color="darkorange")
plt.axvline(df["Age"].median(), color="red", linestyle="dashed", linewidth=2, label=f"Median Age: {df['Age'].median()}")
plt.axvline(df["Age"].mean(), color="green", linestyle="dashed", linewidth=2, label=f"Mean Age: {df['Age'].mean():.1f}")
plt.xlabel("Ya")
plt.ylabel("Milyaderlrin Say")
plt.title("Milyarderlin Ya Paylanmas")
plt.legend()
plt.show()
```

```python
### 3. Cinsiyyt Paylanmas
plt.figure(figsize=(6, 6))
ax = sns.countplot(x=df['Gender'], palette='coolwarm')

# Annotate each bar with its count
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='bottom', fontsize=12, fontweight='bold')

plt.xlabel("Cinsiyyt")
plt.ylabel("Say")
plt.title("Milyaderlrin Cinsiyyt Paylanmas")
plt.show()
```

```python
# Pie chart cinsiyyt qrafikas
plt.figure(figsize=(6, 6))
gender_counts = df['Gender'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=["skyblue", "lightcoral"], startangle=140)
plt.title("Milyaderlrin Cinsiyyt Paylanmas (Pie Chart)")
plt.show()
```

```python
### 4. Top 10 Milyader Net srvtin gr ###
top_10_billionaires = df.nlargest(15, "Net Worth (Billion $)")
plt.figure(figsize=(12, 6))
sns.barplot(y=top_10_billionaires["Name"], x=top_10_billionaires["Net Worth (Billion $)"], palette="coolwarm")
plt.xlabel("Net Srvt (Milyard $)")
plt.ylabel("Milyarder")
plt.title("Net Srvtin gr top 15 milyaderlr")
plt.gca().invert_yaxis()
plt.show()
```

```python
### 5. Top 15 Srvt Mnbyi ###
plt.figure(figsize=(10, 6))
source_counts = df["Source"].value_counts().head(15)
sns.barplot(y=source_counts.index, x=source_counts.values, palette="viridis")
plt.xlabel("Say")
plt.ylabel("Srvt mnbyi")
plt.title("Top 15 Srvt Mnbyi")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
top_industries = df['Industries'].value_counts().head(15)
sns.barplot(y=top_industries.index, x=top_industries.values, palette='magma')
plt.xlabel("Milyarder Say")
plt.ylabel("ndustriya")
plt.title("Milyader sayna gr top 15 industriya")
plt.show()
```

## Milyader sayna gr top 5 d olan lklrin mumi analizi

```python
# Function for top N analysis
def plot_top_n(data, column, title, n=15):
    plt.figure(figsize=(12, 6))
    top_n = data[column].value_counts().head(n)
    sns.barplot(x=top_n.values, y=top_n.index, palette="viridis")
    plt.title(f'Top {n} {title}', fontsize=14)
    plt.xlabel('Count')
    plt.ylabel(title)
    plt.show()
```

```python
top_countries = df['Country'].value_counts().head(5).index.tolist()
for country in top_countries:
    country_df = df[df['Country'] == country]
    print(f"\n {country}: analizi")
    
    # Gender distribution
    plt.figure(figsize=(8, 4))
    sns.countplot(x='Gender', data=country_df, palette='coolwarm')
    plt.title(f'Cinsiyyt Paylanmas {country}', fontsize=12)
    plt.show()
    
    # Industries
    plot_top_n(country_df, 'Industries', f'ndustriya in {country}', 10)
    
    # Age distribution
    plt.figure(figsize=(10, 5))
    sns.boxplot(x='Age', data=country_df, color='skyblue')
    plt.title(f'Ya Paylanmas {country}', fontsize=12)
    plt.show()
    
    # Top sources
    plot_top_n(country_df, 'Source', f' {country} da Srvt Mnbyi', 10)
```
